### Spliting Actor Name 

```python
Actors = Movie_Data.cast.str.split("|",expand=True)  ## Spliting Actors name and making new Data
Actors
```

```python
Actors = Actors.stack().reset_index(level=1,drop=True).to_frame()
Actors
```

### Giving and Column name to new Actors column

```python
Actors.columns = ['Actors_names']
```

### Merging the Actors Data to Movie_Data Set

```python
Actors = Actors.merge(Movie_Data[["title","revenue_musd",
                                  "vote_average","popularity"]],
                      how = 'left', left_index=True, right_index=True)
```

### Checking Top 10 Actors with their Value count

```python
Actors.Actors_names.value_counts().head(10)
```

```python
Actors.Actors_names.value_counts().head(10).plot(kind = 'bar')
```

### Grouping Actors By their Revenue, Rating, Popularity and Movies

```python
Actors_info = Actors.groupby('Actors_names').agg(Total_Revenue = ('revenue_musd', 'sum'),
                                            Mean_Revenue = ('revenue_musd', 'mean'),
                                            Mean_Votes = ('vote_average', 'mean'),
                                            Mean_Popularity = ('popularity', 'mean'),
                                            Total_Movies = ('title', 'count'))
```

### Actors Sorted Based on Mean_Revenue, Total_Movies & Mean_Popularity

```python
Actors_info.sort_values('Mean_Revenue', ascending = False).head()
```

```python
Actors_info.sort_values('Total_Movies', ascending = False).head()
```

```python
Actors_info.sort_values('Mean_Popularity', ascending = False).head()
```
