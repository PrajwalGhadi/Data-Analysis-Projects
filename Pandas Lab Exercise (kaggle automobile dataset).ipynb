**How many cars have horsepower greater than 100?**

```python
count = 0
for i in range(len(data.horsepower)):
    if data.horsepower[i] > 100:
        count += 1
print(f'{count} Cars have Greater than 100 Horsepower')
```

**How many hatchback cars are in the dataset ?**

```python
count = 0
for i in range(len(data.body_style)):
    if data.body_style[i] == 'hatchback':
        count += 1
print(f'{count} Hatchback Cars in Dataset')
```
