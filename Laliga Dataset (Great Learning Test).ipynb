### Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences.
### Using the same function, find the team which has maximum and minimum goal difference.

```python
def Goal_diff_count():
    Football['Goal_diff_count'] = abs(Football['GoalsFor'] - Football['GoalsAgainst'])
    print('Teams with their Goal Difference are:\n',Football.Team, '-', Football.Goal_diff_count)
    
#     print('Maximum Goal Difference is:', Football.Goal_diff_count.max())
#     print('Minimum Goal Difference is:', Football.Goal_diff_count.min())

# for i in range(len(Football)):
#     if Football.Goal_diff_count.min():
#         print(Football.Goal_diff_count[i])
        
    Max_Goal = Football.Goal_diff_count.max()
    Min_Goal = Football.Goal_diff_count.min()
    
    for i in range(len(Football)):
        if Football.Goal_diff_count[i] == Max_Goal:
            print(f'Team with Maximum Goals is: {Football.Team[i]} with {Football.Goal_diff_count[i]} Goals')
        if Football.Goal_diff_count[i] == Min_Goal:
            print(f'Team with Minimum Goals is: {Football.Team[i]} with {Football.Goal_diff_count[i]} Goals')

            
Goal_diff_count()
```

### Create a new column with name “Winning Percent” and append it to the data set (5 points)

### Percentage of Winning = (GamesWon / GamesPlayed)*100

### If there are any numerical error, replace it with 0%

### Print the top 5 teams which has the highest Winning percentage

```python
Football['Winning_Percent'] = round(Football['GamesWon'] / Football['GamesPlayed'] * 100)
```

```python
Football = Football.sort_values(ascending = False, by = 'Winning_Percent')
Football['Winning_Percent'].fillna(value=0)
print('Top 5 teams which has the Highest Winning Percentage are:\n', Football.Team.head(5))
```
